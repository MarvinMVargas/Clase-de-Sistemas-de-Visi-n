```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolution(img,py,px,krnl):
    px = px+1
    py = py+1
    sum = 0 
    for i in range(3):
        for j in range(3):
            newval = img[py-i,px-j] * krnl[i][j]    
            sum = sum + newval
           
    '''if sum > 255:
        sum = 255
    elif sum < 0:
        sum = 0'''

    return sum   
```

```python
img = cv.imread('face.jpg')
img = cv.copyMakeBorder(img,1,1,1,1,cv.BORDER_CONSTANT,None,[0,0,0])
img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
nimg = np.asarray(img, dtype = 'int32')
plt.imshow(nimg, cmap = 'gray')
```

```python
negativeKrnl = [[-1,-1,-1],[-1,8,-1],[-1,-1,-1]]
hog = [[0,-1,0],[-1,0,1],[0,1,0]]

h,w = img.shape
laplaceImg = np.zeros((h,w))

for i in range(1,h-1):
    for j in range(1,w-1):
        laplaceImg[i,j] = convolution(nimg,i,j,hog)  

plt.imshow(laplaceImg, cmap='gray', vmin= 0, vmax=255)
```
